### 3、get请求高级用法/

```python
#HTTP默认的请求方法就是GET
    # * 没有请求体
    # * 数据必须在1K之内！
    #* GET请求数据会暴露在浏览器的地址栏中
    
#GET请求常用的操作：
    #1. 在浏览器的地址栏中直接给出URL，那么就一定是GET请求
    #2. 点击页面上的超链接也一定是GET请求
    #3. 提交表单时，表单默认使用GET请求，但可以设置为POST
```

```python
#1）基本请求的代码
r=requests.get("http://httpbin.org/get")
```

```python
print(r.text)#返回相应的请求信息，包括请求头、URL、IP等信息
```

```python
#2）GET请求的参数
#GET请求的参数放在url的问号后面，以键值对形式传参.
#Params:
#Get方法常见的形式：https://ip:port/xxxx?xx=xx,如果访问带有参数的Url，可以传入一个Dict。
```

```python
#这种网址如何处理？带参数http://httpbin.org/get?name=germey&age=22
```

```python
#问题：get请求如果附加额外信息，如何添加？比如：添加name是germey,age是22，如何构造请求链接？
#解决方案：利用 params参数即可
data={'name':'germey','age':'22'}
r=requests.get('http://httpbin.org/get',params=data)
```

```python
print(r.text)
```

```python
data={'name':'MIke','age':'32'}
r=requests.get('http://httpbin.org/get?name=germey&age=22',params=data)
```

```python
print(r.text)
```

对比两个，会发现data内容作为参数传入到网页内

有很多网站，在其访问web资源时，必须设置一些HTTP请求头，如User-Agent、Host、Cookie等，否则网站服务器会限制访问这些web资源。

```python
#Header：
#也是传入一个Dict作为参数：
```

```python
    #如果是普通网页，以上方法就可以了。如果网页存在反扒的情况，需要header的设置，在请求头内将自己伪装成浏览器，否则百度不会正常返回页面内容 。
```

```python
response=requests.get('https://www.baidu.com/s?wd=python&pn=1',
                     #请求头信息
                      headers={
                        'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.75 Safari/537.36',
                      })

response.encoding='utf-8'
print(response.text)
```

```python
#或者
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.75 Safari/537.36',}
data={'wd':'python','pn':1}
response=requests.get('https://www.baidu.com/s?wd=python&pn=1',
                     #请求头信息
                      headers={
                        'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.75 Safari/537.36',
                      })
response=requests.get('https://www.baidu.com/s',params=data,headers=headers)
response.encoding='utf-8'
print(response.text)
```

```python
with open('baidu.html','w+') as f:
            f.write(response.text)#使用OPEN
```

### get方法指定的url不仅可以是网页，还可以是任何二进制文件，如png图像、pdf文档等，但是二进制由于其编码问题，
### 显示都是乱码，所以需要调用response.content属性获得bytes形式的数据。

```python
#以Github站点图标为例：
r=requests.get('https://github.com/favicon.ico')
print(r.status_code)
print(type(r.text))
print(type(r.content))
print(r.text)#出现乱码，因为图片是二进制
```

```python
#以Github站点图标为例：
r=requests.get('https://ns-strategy.cdn.bcebos.com/ns-strategy/upload/fc_big_pic/part-00431-1339.jpg')
print(r.status_code)
print(r.content)#出现乱码，因为图片是二进制,使用content方法以二进制形式展示
```

```python
with open('favicon.ico','wb') as f:
            f.write(r.content)#使用OPEN的方法，以二进制写的形式打开，可以向文件里写二进制数据。
```

## 4、post请求

```python
#除了get请求，还有POST请求。示例如下：以请求http://httpbin.org/post为例，该网站可以判断如果请求是post方式，就把相关请求信息返回。
import requests
data = {'name': 'germey', 'age': '22'}
response = requests.post("http://httpbin.org/post", data=data)
print(response.text)
```

```python
#可以发现，成功获得返回结果，其中form部分就是提交的数据，这就证明POST请求成功发起。
```

## 5、响应

```python
#发送HTTP请求后，get或者post方法会返回响应（response对象）。前面，使用text和content获取了响应的内容。除此之外，还有
#状态码、响应头、Cookie。
```

```python
import requests
header={'User-Agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.106 Mobile Safari/537.36'}
response = requests.get('http://www.jianshu.com',headers=header)
print(type(response.status_code), response.status_code)
print(type(response.headers), response.headers)
print(type(response.cookies), response.cookies)
print(type(response.url), response.url)
print(type(response.history), response.history)
```

## 状态码判断

```python
import requests
header={'User-Agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.106 Mobile Safari/537.36'}
response = requests.get('http://www.jianshu.com',headers=header)
if response.status_code != 200:
    exit()  
else:
    print('Request Successfully')
```

```python
import requests
header={'User-Agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.106 Mobile Safari/537.36'}
r= requests.get('http://www.jianshu.com',headers=header)
if not r.status_code == requests.codes.ok:
    print('failed') 
else:
    print('Request Successfully')
```
