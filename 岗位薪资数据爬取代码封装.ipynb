```python
#定义获取一页的详细页URL的函数,输入某一页URL,输出详细页网址
def get_links(url):
    res=requests.get(url,headers=headers,cookies=cookie)
    res.encoding='utf-8'
    re=res.text
    html = etree.HTML(re)
    url_sons=html.xpath('//div[@class="items"]/a/@href') 
    return(url_sons)
```

```python
#定义获取一详细网页信息的函数，输入详细页网址，输出该页数据
def get_info(url):
    res_son=requests.get(url,headers=headers)
    res_son.encoding='utf-8'
    html_son = etree.HTML(res_son.text)
    gang=html_son.xpath('//div[@class="jt"]/p/text()')
    didian=html_son.xpath('//div[@class="jt"]/em/text()')
    time=html_son.xpath('//div[@class="jt"]/span/text()')
    money=html_son.xpath('//p[@class="jp"]/text()')
    people=html_son.xpath('//div[@class="jd"]/span/text()')
    name=html_son.xpath('//div[@class="rec"]/a/p[@class="c_444"]/text()')
    xingz=html_son.xpath('//div[@class="rec"]/a/div[@class="at"]/text()')
    dizhi=html_son.xpath('//div[@class="rec"]/a/span/text()')
    describe=''.join(html_son.xpath('//div[@class="welfare"]/span/text()'))
    work=''.join(html_son.xpath('//div[@class="ain"]/article/p/text()'))
    renz=''.join(html_son.xpath('//div[@class="ain"]/article/p/text()'))
    data={'岗位':gang,'地点':didian,'时间':time,'薪资':money,'人数':people,'公司名':name,'公司性质':xingz,
              '公司地址':dizhi,'职位描述':describe,'职责描述':work,'任职':renz}
    print(data)
```
